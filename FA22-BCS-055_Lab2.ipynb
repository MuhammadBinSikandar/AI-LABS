```python
# Activity 2
def isPalindrome(word):
    temp = word[::-1]
    if temp.capitalize() == word.capitalize():
        return True
    else:
        return False
    
word = input("Enter a word : ")
print("Palindrome" if isPalindrome(word) else "Not Palindrome")

```

```python
# Activity 3
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

rows_A = len(A)
cols_A = len(A[0])
rows_B = len(B)
cols_B = len(B[0])

if cols_A != rows_B:
    print("Number of columns in A must be equal to number of rows in B")
else:
    C = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    for i in range(rows_A):
        for j in range(cols_B):
            sum_product = 0
            for k in range(cols_A):
                sum_product += A[i][k] * B[k][j]
            C[i][j] = sum_product

    print("Resulting Matrix C:")
    for row in C:
        print(row)
```

```python
# Activity 4
def perimeter(listing):
    perimeter = 0
    for i in range(0,len(listing)-1):
        dist = ((listing[i][0]-listing[i+1][0])**2)+((listing[i][1]-listing[i+1][1])**2)**0.5
        perimeter += dist
    perimeter = perimeter + (((listing[0][0]-listing[-1][0])**2)+((listing[0][1]-listing[-1][1])**2)**0.5)
    return perimeter

listing = [[0,0],[1,0],[1,1],[0,1]]
print("Perimeter is :",perimeter(listing))
```

```python
# Activity 5
def symmetricDifference(set1, set2):
    emptySet = set()
    for i in set1:
        if i not in set2:
            emptySet.add(i)
    for i in set2:
        if i not in set1:
            emptySet.add(i)
    return emptySet

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
print(symmetricDifference(set1, set2))
print(set1.symmetric_difference(set2))
print(set2.symmetric_difference(set1))
print(set1 ^ set2)
print(set2 ^ set1)
```

```python
# Graded Lab Task 3
from math import *

def derivative_of_Sin(x, h):
    return (sin(x + h) - sin(x)) / h

x_range = [i / 1000 for i in range(int(-pi * 1000), int(pi * 1000) + 1)]

for x in x_range:
    print(f"x: {x:.5f}, Approx: {derivative_of_Sin(x, 0.001):.5f}, cos(x): {cos(x):.5f}")

for h in [0.01, 0.1]:
    print(f"\nFor h = {h}")
    for x in x_range[::1000]:
        print(f"x: {x:.5f}, Approx: {derivative_of_Sin(x, h):.5f}, cos(x): {cos(x):.5f}")
```
